# Fibonacci Series

```python
n = 10 #input

a = 0 #first number in Fibonacci Series
b = 1 #second number in Fibonacci Series
i = 0 #counter variable

print (a, end=' ')
print (b, end=' ')
while i < n-2:
  c = a + b
  a = b
  b = c
  i = i+1

  print (c, end=' ')
```
